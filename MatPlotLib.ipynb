```python
edad_x =[20,25,30,35,40,45,50,55,60,65]
```

```python
plt.plot(edad_x, salario_cd_y)
plt.title("Salario medio en USD de un ceintífico de datos por edad")
plt.xlabel("edad")
plt.ylabel("salario")
plt.show
```

```python
plt.plot(edad_x, salario_fed_y, label="Front End developer")
plt.plot(edad_x, salario_cd_y, label="Científico de datos")
plt.plot(edad_x, salario_bed_y, label="BackEnd developer")


plt.plot(edad_x, salario_cd_y)
plt.title("Salario medio en USD por edad")
plt.xlabel("edad")
plt.ylabel("salario")
plt.legend()
plt.show()
```

```python
plt.plot(edad_x, salario_fed_y, label="Front End developer", linewidth=3, marker='.', markersize=20, linestyle='--', color='#d62728')
plt.plot(edad_x, salario_cd_y, label="Científico de datos")
plt.plot(edad_x, salario_bed_y, label="BackEnd developer")
plt.grid(True)

plt.plot(edad_x, salario_cd_y)
plt.title("Salario medio en USD por edad")
plt.xlabel("edad")
plt.ylabel("salario")
plt.legend()
plt.show()
```

```python
plt.plot(edad_x, salario_fed_y, label="Front End developer", linewidth=3, marker='.', markersize=20, linestyle='--', color='#d62728')
plt.plot(edad_x, salario_cd_y, label="Científico de datos")
plt.plot(edad_x, salario_bed_y, label="BackEnd developer")
plt.grid(True)
import numpy as np
salario_medio= np.sum([salario_cd_y, salario_fed_y, salario_bed_y],axis=0)/3

plt.plot(edad_x, salario_medio, label = "Salario medio", linestyle='--', color = '#000000')
plt.title("Salario medio en USD por edad")
plt.xlabel("edad")
plt.ylabel("salario")
plt.legend()
plt.show()
```

```python
plt.plot(edad_x, salario_bed_y, label="BackEnd developer")
plt.grid(True)
import numpy as np
salario_medio_y= np.sum([salario_cd_y, salario_fed_y, salario_bed_y],axis=0)/3

plt.plot(edad_x, salario_medio, label = "Salario medio", linestyle='--', color = '#000000')

plt.fill_between(edad_x,salario_bed_y,salario_medio_y,
                interpolate=True, alpha=0.33, label = 'Between')
plt.title("Salario medio en USD por edad")
plt.xlabel("edad")
plt.ylabel("salario")
plt.legend()
plt.show()
```

```python
plt.plot(edad_x, salario_bed_y, label="BackEnd developer")
plt.grid(True)
import numpy as np
salario_medio_y= np.sum([salario_cd_y, salario_fed_y, salario_bed_y],axis=0)/3

plt.plot(edad_x, salario_medio, label = "Salario medio", linestyle='--', color = '#000000')

plt.fill_between(edad_x,salario_bed_y,salario_medio_y, where=(salario_bed_y > salario_medio_y),
                interpolate=True, alpha=0.33, label = 'Between')
plt.fill_between(edad_x,salario_bed_y,salario_medio_y, where=(salario_bed_y <= salario_medio_y),
                interpolate=True, color = 'yellow', alpha=0.33, label = 'Between')
plt.title("Salario medio en USD por edad")
plt.xlabel("edad")
plt.ylabel("salario")
plt.legend()
plt.show()
```
